```python
#Dropping the empty rows using dropna()
Coal_cons = Coal_consumption.dropna()
Coal_cons.head(10)
```

```python
print(Coal_cons.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Coal_cons.columns = Coal_cons.iloc[0]
Coal_cons.reset_index(drop=True, inplace=True)
Coal_cons = Coal_cons.drop(axis=1, index=0)
Coal_cons.head(15)
```

```python
#dropping the last 3 columns which have calculated data that and is't part of the yearly energy consumption
Coal_cons = Coal_cons.iloc[:, :-3]
Coal_cons.head()
```

```python
#Changing the first column's name from : "Exajoules" to "Country"
Coal_cons.rename(columns={"Exajoules": "Country"}, inplace=True)

#Reviewing the columns names
Coal_cons.columns
```

```python
#Changing the type of the years column headers from float to integer
Coal_cons.columns = [int(col) if isinstance(col, float) else col for col in Coal_cons.columns]
Coal_cons.columns
```

```python
cleaned_rows = []
rows = Coal_cons.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # Extracting the region name when finding a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assigning this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx) :
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1  # move start index to next group
```

```python
# Create new DataFrame from cleaned list
Coal_cons = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Getting all year columns
year_columns = [col for col in Coal_cons.columns if isinstance(col, int)]
```

```python
long_coal = pd.melt(
    Coal_cons,
    id_vars=['Country', 'Region'],          # keeping country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Coal Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_coal.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_coal_consumption.csv', index=False)
```
