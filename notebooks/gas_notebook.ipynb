```python
#Dropping the empty rows using dropna()
Gas_cons = Gas_consumption.dropna()
Gas_cons.head(10)
```

```python
print(Gas_cons.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Gas_cons.columns = Gas_cons.iloc[0]
Gas_cons.reset_index(drop=True, inplace=True)
Gas_cons = Gas_cons.drop(axis=1, index=0)
Gas_cons.head(15)
```

```python
#dropping the last 3 columns which have calculated data that and is't part of the yearly energy consumption
Gas_cons=Gas_cons.iloc[:, :-3]
Gas_cons.head()
```

```python
#Changing the first column's name from : "Exajoules" to "Country"
Gas_cons.rename(columns={"Exajoules": "Country"}, inplace=True)
Gas_cons.columns
```

```python
#Changing the type of the years column headers from float to integer
Gas_cons.columns = [int(col) if isinstance(col, float) else col for col in Gas_cons.columns]
Gas_cons.columns
```

```python
cleaned_rows = []
rows = Gas_cons.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # Extracting the region name when finding a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assigning this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx) :
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1  # move start index to next group
```

```python
# Create new DataFrame from cleaned list
Gas_cons = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Getting all year columns
year_columns = [col for col in Gas_cons.columns if isinstance(col, int)]
```

```python

long_gas = pd.melt(
    Gas_cons,
    id_vars=['Country', 'Region'],          # keeping country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Gas Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_gas.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_gas_consumption.csv', index=False)
```
