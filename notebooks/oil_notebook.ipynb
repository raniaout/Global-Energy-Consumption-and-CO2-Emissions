```python
#Dropping the empty rows using dropna()
Oil_cons = Oil_consumption.dropna()
Oil_cons.head(10)
```

```python
print(Oil_cons.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Oil_cons.columns = Oil_cons.iloc[0]
Oil_cons.reset_index(drop=True, inplace=True)
Oil_cons= Oil_cons.drop(axis=1, index=0)
Oil_cons.head(15)
```

```python
#dropping the last 3 columns 
Oil_cons = Oil_cons.iloc[:, :-3]
```

```python
#Changing the first column's name to "Country"
Oil_cons.rename(columns={"Exajoules": "Country"}, inplace=True)
Oil_cons.columns
```

```python
#Changing the type of the years column headers from float to integer
Oil_cons.columns = [int(col) if isinstance(col, float) else col for col in Oil_cons.columns]
Oil_cons.columns
```

```python
cleaned_rows = []
rows = Oil_cons.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # When we hit a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assign this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx):
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1  # move start index to next group
```

```python
# Create new DataFrame from cleaned list
Oil_cons = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Get all year columns (they are usually strings or ints starting from 1965)
year_columns = [col for col in Oil_cons.columns if isinstance(col, int)]
```

```python
long_oil = pd.melt(
    Oil_cons,
    id_vars=['Country', 'Region'],          # keep country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Oil Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_oil.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_oil_consumption.csv', index=False)
```
