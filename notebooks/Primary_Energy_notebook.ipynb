```python
#Dropping the empty rows using dropna()
Primary_Energy_consumption = Energy_consumption.dropna()

#Reviewing the table
Primary_Energy_consumption.head(10)
```

```python
print(Primary_Energy_consumption.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Primary_Energy_consumption.columns = Primary_Energy_consumption.iloc[0]
Primary_Energy_consumption.reset_index(drop=True, inplace=True)
Primary_Energy_consumption= Primary_Energy_consumption.drop(axis=1, index=0)
Primary_Energy_consumption.head(15)
```

```python
#dropping the last 3 columns which have calculated data that and is't part of the yearly energy consumption
Primary_Energy_consumption=Primary_Energy_consumption.iloc[:, :-3]
Primary_Energy_consumption.head()
```

```python
#Changing the first column's name to "Country"
Primary_Energy_consumption.rename(columns={"Exajoules": "Country"}, inplace=True)

Primary_Energy_consumption.columns
```

```python
#Changing the type of the years column headers from float to integer
Primary_Energy_consumption.columns = [int(col) if isinstance(col, float) else col for col in Primary_Energy_consumption.columns]
Primary_Energy_consumption.columns
```

```python
cleaned_rows = []
rows = Primary_Energy_consumption.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assign this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx):
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1 
```

```python
Primary_Energy_consumption = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Get all year columns 
year_columns = [col for col in Primary_Energy_consumption.columns if isinstance(col, int)]
```

```python
long_energy = pd.melt(
    Primary_Energy_consumption,
    id_vars=['Country', 'Region'],          # keep country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Energy Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_energy.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_energy_consumption.csv', index=False)
```
