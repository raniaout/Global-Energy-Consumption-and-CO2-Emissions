```python
#Dropping the empty rows using dropna()
Renewables_cons = Renewables_consumption.dropna()

#Reviewing the table
Renewables_cons.head(7)
```

```python
#Checking if there are still empty rows
print(Renewables_cons.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Renewables_cons.columns = Renewables_cons.iloc[0]
Renewables_cons.reset_index(drop=True, inplace=True)
Renewables_cons = Renewables_cons.drop(axis=1, index=0)
Renewables_cons.head()
```

```python
#dropping the last 3 columns which have calculated data and  isn't part of the yearly energy consumption
Renewables_cons=Renewables_cons.iloc[:, :-3]
Renewables_cons.head()
```

```python
#Changing the first column's name to "Country"
Renewables_cons.rename(columns={"Exajoules (input-equivalent)": "Country"}, inplace=True)

#Reviewing the columns names
Renewables_cons.columns
```

```python
#Changing the type of the years column headers from float to integer
Renewables_cons.columns = [int(col) if isinstance(col, float) else col for col in Renewables_cons.columns]
Renewables_cons.columns
```

```python
cleaned_rows = []
rows = Renewables_cons.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # Extracting the region name when finding a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assigning this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx) :
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1 
```

```python
# Create new DataFrame from cleaned list
Renewables_cons = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Getting all year columns
year_columns = [col for col in Renewables_cons.columns if isinstance(col, int)]
```

```python

long_Renewables = pd.melt(
    Renewables_cons,
    id_vars=['Country', 'Region'],          # keeping country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Renewables Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_Renewables.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_Renewables_consumption.csv', index=False)
```
