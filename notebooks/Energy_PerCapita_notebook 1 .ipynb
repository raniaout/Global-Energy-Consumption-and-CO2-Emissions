```python
# import pandas and load the sheet
import pandas as pd
Energy_Cons_capita = pd.read_excel("C:/Users/pc/Desktop/+++/2ème année Génie Chimique/S3/Raffinage du pétrole (4)/séminaire/sources utilisées/done/bp-stats-review-2022-all-data.xlsx", sheet_name='Primary Energy - Cons capita', header=None)
```

```python
#Dropping the empty rows using dropna()
Energy_Cons_capita = Energy_Cons_capita.dropna()
Energy_Cons_capita.head(10)
```

```python
print(Energy_Cons_capita.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Energy_Cons_capita.columns = Energy_Cons_capita.iloc[0]
Energy_Cons_capita.reset_index(drop=True, inplace=True)
Energy_Cons_capita= Energy_Cons_capita.drop(axis=1, index=0)
Energy_Cons_capita.head(15)
```

```python
#dropping the last 3 columns which have calculated data that and is't part of the yearly energy consumption
Energy_Cons_capita = Energy_Cons_capita.iloc[:, :-2]
Energy_Cons_capita.head()
```

```python
#Changing the first column's name to "Country"
Energy_Cons_capita.rename(columns={"Gigajoule per capita": "Country"}, inplace=True)
Energy_Cons_capita.columns
```

```python
#Changing the type of the years column headers from float to integer
Energy_Cons_capita.columns = [int(col) if isinstance(col, float) else col for col in Energy_Cons_capita.columns]
Energy_Cons_capita.columns
```

```python
cleaned_rows = []
rows = Energy_Cons_capita.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # When we hit a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assign this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx):
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1  # move start index to next group
```

```python
# Create new DataFrame from cleaned list
Energy_Cons_capita = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Get all year columns
year_columns = [col for col in Energy_Cons_capita.columns if isinstance(col, int)]
```

```python
long_energy = pd.melt(
    Energy_Cons_capita,
    id_vars=['Country', 'Region'],          # keep country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Energy Consumption per Capita (GJ)'    # name for the values
)
```
