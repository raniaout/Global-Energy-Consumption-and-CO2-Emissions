```python
#Adding a new column 'Energy Type' to all energy consumption dataframes (so all except the co2_emissions and per capita consumption)
Primary_consumption['Energy Type'] = "Primary"
Coal_consumption['Energy Type'] = 'Coal'
Gas_consumption['Energy Type'] = 'Gas'
Oil_consumption['Energy Type'] = 'Oil'
Hydro_consumption['Energy Type'] = 'Hydro'
Nuclear_consumption['Energy Type'] = 'Nuclear'
Renewables_consumption['Energy Type'] = 'Renewables'
```

```python
#Unifying the name of the energy consumption column
for df in [Primary_consumption, Coal_consumption, Gas_consumption, Oil_consumption, Hydro_consumption, Nuclear_consumption, Renewables_consumption] : 
    df.rename(columns={'Renewables Consumption (EJ)' : 'Energy Consumption (EJ)', 'Nuclear Consumption (EJ)' :'Energy Consumption (EJ)', 'Hydro Consumption (EJ)': 'Energy Consumption (EJ)', 'Coal Consumption (EJ)': 'Energy Consumption (EJ)', 'Gas Consumption (EJ)': 'Energy Consumption (EJ)', 'Oil Consumption (EJ)' : 'Energy Consumption (EJ)'}, inplace=True)
```

```python
#Concatenating them into one fact table 
Energy_Fact = pd.concat([Coal_consumption, Gas_consumption, Oil_consumption, Hydro_consumption, Nuclear_consumption, Renewables_consumption, Primary_consumption], ignore_index=True)
```

```python
#Creating the dimension tables
#We'll have 3 dimension tables : dim_country , dim_year and dim_energy_type
Dim_Year = pd.DataFrame({'Year' : sorted(Energy_Fact['Year'].unique())}) 
Dim_Year['YearID'] =  range(965, 965+ len(Dim_Year))
Dim_Year.head(20)
```

```python
Dim_Country = Energy_Fact[['Country', 'Region']].drop_duplicates()
```

```python
Dim_Country['CountryID'] = range(101, 101 + len(Dim_Country))
```

```python
Dim_Country.info()
```

```python
Dim_Energy_Type = pd.DataFrame({'Energy Type' : Energy_Fact['Energy Type'].drop_duplicates()})
Dim_Energy_Type['Energy Type Code'] = range(201 , 201+len(Dim_Energy_Type))
```

```python
#Merging the Energy_Fact table with the dimension tables

Energy_Fact = Energy_Fact.merge(Dim_Country, on=['Country'] )
Energy_Fact = Energy_Fact.merge(Dim_Year, on=['Year'] )
Energy_Fact = Energy_Fact.merge(Dim_Energy_Type, on=['Energy Type'] )
```

```python
# Adding an ID column to the energy fact table
Energy_Fact['ConsumptionID'] = range(301, 301 + len(Energy_Fact))
```
