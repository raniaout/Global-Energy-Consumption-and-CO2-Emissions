```python
#Dropping the empty rows using dropna()
Nuclear_cons = Nuclear_consumption.dropna()
Nuclear_cons.head(10)
```

```python
print(Nuclear_cons.isnull().sum())
```

```python
#Setting the first row as the columns header and then dropping that row 
Nuclear_cons.columns = Nuclear_cons.iloc[0]
Nuclear_cons.reset_index(drop=True, inplace=True)
Nuclear_cons = Nuclear_cons.drop(axis=1, index=0)
Nuclear_cons.head(15)
```

```python
#dropping the last 3 columns which have calculated data that and is't part of the yearly energy consumption
Nuclear_cons=Nuclear_cons.iloc[:, :-3]
Nuclear_cons.columns
```

```python
#Changing the first column's name to "Country"
Nuclear_cons.rename(columns={"Exajoules (input-equivalent)": "Country"}, inplace=True)
Nuclear_cons.columns
```

```python
#Changing the type of the years column headers from float to integer
Nuclear_cons.columns = [int(col) if isinstance(col, float) else col for col in Nuclear_cons.columns]
Nuclear_cons.columns
```

```python
cleaned_rows = []
rows = Nuclear_cons.reset_index(drop=True)

start_idx = 0

for idx, row in rows.iterrows():
    country = row['Country']
    # Extracting the region name when finding a total row
    if str(country).startswith("Total "):
        region = country.replace("Total ", "").strip()
        # Assigning this region to all rows from start_idx to idx - 1
        for j in range(start_idx, idx) :
            country_row = rows.iloc[j].copy()
            country_row['Region'] = region
            cleaned_rows.append(country_row)
        start_idx = idx + 1  # move start index to next group
```

```python
# Create new DataFrame from cleaned list
Nuclear_cons = pd.DataFrame(cleaned_rows).reset_index(drop=True)
```

```python
# Getting all year columns
year_columns = [col for col in Nuclear_cons.columns if isinstance(col, int)]
```

```python
long_nuclear = pd.melt(
    Nuclear_cons,
    id_vars=['Country', 'Region'],          # keeping country and region
    value_vars=year_columns,                # years become rows
    var_name='Year',                        # new column name for former columns
    value_name='Nuclear Consumption (EJ)'    # name for the values
)
```

```python
#Save this Dataset 
long_nuclear.to_csv('C:/Users/pc/Desktop/Formation Data Analyst/Projet/New Dataset/total_yearly_nuclear_consumption.csv', index=False)
```
